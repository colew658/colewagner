## Research Goals and Project Importance

The goal of this analysis was to predict the price of a used car using a variety of features such as title status, engine cylinders, mileage, and many more. To do this, I created both random forest and XG boost models and compared their performance. In addition, a secondary research goal was to determine the most important features in price prediction.

Being able to accurately predict the price of a used car is valuable to many different groups of people, a few of which are highlighted below.

**Used car dealers** need to know the values of used cars so they can make smart decisions when purchasing vehicles at auction. Car auctions are typically settings in which manufacturers and wholesale buyers will sell overstock cars to dealers at a discounted price. However, if a dealer isn't aware of the true value of a car, they would have no idea how much they could comfortably bid on said car. Without this cruicial information, dealers would quickly go out of business. 

On the opposite side, **retail buyers** in the market for a used car need to have a good estimate of what a given car is worth as well. Uneducated consumers can be taken advantage of when trying to buy a used car because they don't understand what their desired car is worth. They could overpay by hundreds or even thousands of dollars. To prevent this, retail buyers must do extensive research to estimate the value of the car they want before showing up to the dealership. This model could certainly aid them in that pursuit.

**The auto finance industry** is also very interested in the value of a used car. This most often comes into play when these organizations must decide on the residual value of a vehicle that is being returned after a lease. After an owner returns a leased vehicle, they have the option to buy the car at its current residual value. Data scientists working in the auto finance industry often build models to predict this residual value. This allows car manufacturers to sell more cars while also making dealers more profitable. Setting this price at a fair value is paramount to both of the former objectives.

**Insurance companies** need to know the value of their customer' vehicles to be able to pay out a fair amount in the event of an accident that totals a vehicle. In this case, the insurance company pays the customer the amount the car was worth just before it crashed. Being able to calculate this value accurately ensures that insurance companies can remain profitable.

## Data Summary

### Craigslist Data

The first dataset used in this analysis was scraped from Craigslist. It contains 426,880 used car listings, with 26 variables for each listing. These variables include Craigslist-specific information such as website links and the date and description of the listing. There are also a variety of variables describing the location of the listing, some being as broad as what state the listing is in, all the way down to the latitude and longitude coordinates of the listing. Most importantly, though, the dataset contains a plethora of information on the cars themselves. This includes key identifiers like year, make, and model. There is also information on the vehicles drivetrain, condition, history, mileage, and even paint color. 

Data link: https://www.kaggle.com/datasets/austinreese/craigslist-carstrucks-data

### EPA Data

The other dataset used below is provided by the Environmental Protection Agency (EPA). The EPA collects data on the fuel efficiency of vehicles to ensure that they meet the organization's standards. This dataset has many niche variables that are not relevant to this study, but it does have a lot of valuable information as well. Specifically, it contains information on city and highway MPG, annual petroleum consumption, fuel type, engine and transmission specifications, and the same key identifiers as above. It features 47,365 rows (vehicles), with 84 variables for each vehicle.

Data link: https://www.fueleconomy.gov/feg/download.shtml

### Dealing with Missing Values

```python
# check for NaN values
((relevantdata.isna().sum())/len(relevantdata)).to_frame()
```

### Cylinders

There are almost no missing values in the cylinders column from the EPA data, while over 40% of the cylinders column is missing from the craigslist data. So, I will drop the craigslist cylinders and remove all rows with missing cylinders from the EPA. This is because imputing the cylinders could result in nonsensical values (such as a car with an engine displacement of 6 liters but only 4 cylinders).

```python
relevantdata.dropna(subset=['cylinders_x'], inplace=True)
relevantdata.drop("cylinders_y", axis=1, inplace=True)
```

```python
# check for NaN values
((relevantdata.isna().sum())/len(relevantdata)).to_frame()
```

### Drive Type

Similar to cylinders, this is a duplicate variable. Because the craiglist version of this variable has many more missing values than the EPA version, I will drop the craigslist version and impute the EPA version with the most common value.

```python
relevantdata.drive_x.value_counts().sort_values().plot(kind = 'barh')
```

```python
relevantdata.drive_x.fillna("Front-Wheel Drive", inplace=True)
relevantdata.drop("drive_y", axis=1, inplace=True)
```

```python
relevantdata.guzzler.fillna("N", inplace=True)
relevantdata.sCharger.fillna("N", inplace=True)
relevantdata.tCharger.fillna("N", inplace=True)
relevantdata.startStop.fillna("N", inplace=True)
```

The price variable must be log transformed to mitigate the effect of outliers and to ensure the model never gives negative price predictions.

### Model Specific Preprocessing

```python
# one hot encode categorical variables
onehot_data = pd.get_dummies(relevantdata.drop(["manufacturer", "model"], axis=1), dtype="float64")
# add model and manufacturer to one hot encoded data
full_modeldata = pd.concat([relevantdata[["manufacturer", "model"]], onehot_data], axis=1)
```

```python
# normalize numeric predictors
scaler = StandardScaler()
columns_to_normalize = ['barrels08','city08', 'comb08','displ', 'fuelCost08', 'highway08', 'UCity', 'UHighway', 'year', 'youSaveSpend', 'odometer']
onehot_data[columns_to_normalize] = scaler.fit_transform(onehot_data[columns_to_normalize])
```

## Data Modeling

```python
# 70/30 train/test split 
X_train, X_test, y_train, y_test = train_test_split(onehot_data.drop(["price", "log_price"], axis=1), onehot_data["log_price"],
                                                    test_size = 0.3, random_state = 0)
```

### XGBoost

```python
# create a parameter grid
xgb_params = {
    "xgb_reg__objective": ["reg:squarederror"],
    'xgb_reg__max_depth': [4, 6, 8],
    'xgb_reg__learning_rate': [0.3],
    'xgb_reg__reg_alpha': [0.1, 1, 5, 10, 25, 50, 100],
    'xgb_reg__reg_lambda': [0.1, 1, 5, 10, 25, 50, 100],
    'xgb_reg__n_estimators': [50, 75, 100, 150]
}

# create xgb pipeline
xgb_pipe = Pipeline(
    steps=[
        ("xgb_reg", XGBRegressor(n_jobs = 7))
    ]
)
```

```python
# fit the model to the training data
xgb_gscv = GridSearchCV(xgb_pipe, xgb_params)
xgb_gscv.fit(X_train, y_train)
```

```python
xgb_gscv.best_params_
```

```python
xgb_pred = xgb_gscv.predict(X_test)
mean_squared_error(y_true=y_test, y_pred=xgb_pred, squared=False)
```

```python
r2_score(y_true=y_test, y_pred=xgb_pred)
```

```python
feat_importances = pd.Series(xgb_gscv.best_estimator_["xgb_reg"].feature_importances_, index=X_train.columns)

fig, ax = plt.subplots()
ax.barh(range(len(feat_importances.nlargest(15))), feat_importances.nlargest(15))
plt.xlabel("Variable Importance")
ax.set_yticks(range(15))
ax.set_yticklabels(feat_importances.nlargest(15).index)
plt.show()
```

### Random Forest

```python
# create a parameter grid
rf_params = {
    'rf_reg__max_depth': [4, 6, 8, 10, 12],
    'rf_reg__ccp_alpha': [0.1, 1, 5, 10, 25, 50, 100],
    'rf_reg__n_estimators': [50, 100, 150, 200]
}

# create xgb pipeline
rf_pipe = Pipeline(
    steps=[
        ("rf_reg", RandomForestRegressor(random_state=1, n_jobs=7))
    ]
)
```

```python
# fit the model to the training data
rf_gscv = GridSearchCV(rf_pipe, rf_params)
rf_gscv.fit(X_train, y_train)
```

```python
rf_gscv.best_params_
```

```python
# make predictions and get RMSE
rf_pred = rf_gscv.predict(X_test)
mean_squared_error(y_true=y_test, y_pred=rf_pred, squared=False)
```

```python
r2_score(y_true=y_test, y_pred=rf_pred)
```

```python
rf_feat_importances = pd.Series(rf_gscv.best_estimator_["rf_reg"].feature_importances_, index=X_train.columns)

fig, ax = plt.subplots()
ax.barh(range(len(rf_feat_importances.nlargest(10))), rf_feat_importances.nlargest(10))
plt.xlabel("Variable Importance")
ax.set_yticks(range(10))
ax.set_yticklabels(rf_feat_importances.nlargest(10).index)
plt.show()
```

### Model Comparison

**Root Mean Square Error**

```python
rmse_comp = pd.DataFrame([["XG Boost", mean_squared_error(y_true=y_test, y_pred=xgb_pred, squared=False), r2_score(y_true=y_test, y_pred=xgb_pred)],
                         ["Random Forest", mean_squared_error(y_true=y_test, y_pred=rf_pred, squared=False), r2_score(y_true=y_test, y_pred=rf_pred)]],
                        columns=["Model", "Root Mean Square Error", "R Squared"])


rmse_comp
```

As we can see from the RMSE and R squared values above, the XG Boost model is much better than the Random Forest. The RMSE for the Random Forest is almost double that of the XG Boost model, so the XG Boost model's prediction are twice as good as the Random Forest's on average. This is related to the disparity in the R squared values between the two modeling frameworks. The Random Forest model only explains about 9.5% of the variance in the test data, while the XG Boost model explains over 70% of the variance in the data. This is further explained by the variable importance plots. The XG Boost model relies on a wide range of variables to make predictions, weighting many variables similarly in terms of importance. The Random Forest, on the other hand, relies almost solely on a car's mileage to make predictions. While this variable is undoubtedly important, it is entirely insufficient to accurately predict the price of a given car. The XG Boost's most important variables are standard 4WD pickup trucks, front wheel drive vehicles, and vehicles with 3 cylinder engines. Each of these variables capture a different niche in the data. 4WD pickups are massively popular as both commercial and personal vehicles. They also hold their value incredibly well. Front-wheel drive cars are typically practical economy cars, as they handle better in snow and rain than rear-wheel drive cars, without being as complicated (and expensive) as all-wheel drive vehicles. Finally, vehicles with 3 cylinder engines are very rare and thus are distinctive as their own category of pricing. All of this shows that the XG Boost model does a much better job at finding unique trends in the data, while the Random Forest is incredibly generic and ineffective.
